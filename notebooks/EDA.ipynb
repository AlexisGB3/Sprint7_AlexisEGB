# Importar librerias

```python
import pandas as pd
import plotly.express as px
```

### Limpieza de datos

```python
# 1. Cargar datos originales
df = pd.read_csv('../data/vehicles_us.csv')

# 2. Limpiar los datos (ejemplos)
df = df.dropna()  # Eliminar filas con valores faltantes
df = df.drop_duplicates()  # Eliminar duplicados
df['price'] = df['price'].astype(float)  # Corregir tipos de datos

# 3. Guardar datos limpios
df.to_csv('../data/vehicles_us_clean.csv', index=False)
```

### Histograma

```python
car_data = pd.read_csv('../data/vehicles_us_clean.csv') # leer los datos
fig = px.histogram(car_data, x="odometer") # crear un histograma
fig.show() # crear grfico de dispersin
```

### Grfico de dispersin

```python
fig = px.scatter(car_data, x="odometer", y="price") # crear un grfico de dispersin
fig.show() # crear grfico de dispersin
```
